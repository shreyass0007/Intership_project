```python

product_sales = df.groupby('Product')['Total_Sales'].sum().sort_values(ascending=False)
print("Products by Total Revenue:")
print(product_sales)
print(f"\nBest Product: {product_sales.idxmax()} with ₹{product_sales.max():,.2f}")
```
